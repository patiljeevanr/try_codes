```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
```

```python
df  = pd.read_csv('train.csv')
```

```python
df.describe()
```

```python
df.drop(['Alley', 'MiscFeature', 'Fence', 'PoolQC', 'Id'], axis = 1, inplace =True)
```

```python
for column in df.columns:
    df[column] = df[column].fillna(df[column].mode()[0])
```

```python
for column in df.columns:
    print(column+": ", sum(pd.isnull(df[column])))
```

```python
Data = pd.get_dummies(df)
```

```python
Data
```

```python
Inputs = Data.drop(['SalePrice'], axis = 1)
```

```python
Target = Data['SalePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(Inputs, Target, test_size = 0.2)
clf = RandomForestRegressor(max_depth=100, random_state=0)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
r2_score(y_pred, y_test)
```

```python
df1 = pd.read_csv('test.csv')
Ids = df1['Id']
df1.drop(['Alley', 'MiscFeature', 'Fence', 'PoolQC', 'Id'], axis = 1, inplace = True)
for column in df1.columns:
    df1[column] = df1[column].fillna(df1[column].mode()[0])
Test_Data = pd.get_dummies(df1)
```

```python
for column in Inputs.columns:
    if column not in Test_Data.columns:
        Test_Data[column] = 0
```

```python
y_test_pred = clf.predict(Test_Data)
```

```python
y_test_pred
```

```python
results = pd.DataFrame({'Id':Ids, 'SalePrice':y_test_pred})
results.to_csv('Results.csv', index = None)
```
